```python
import numpy as np
import matplotlib.pyplot as plt
import pickle

import matplotlib as mpl
plt.rcParams.update({
    "text.usetex": True})

plt.style.use('seaborn-paper')
```

```python
with open('data.pickle', 'rb') as f:
    data = pickle.load(f)
```

## Plotting Dynamics

```python
# --------------Params-------------- #
size_of_mem = 3
K = 75
list_eps = [0.001, 0.01, 0.03, 0.1]
time_step = 0.2
# ---------------------------------- #

# plotting
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(labelcolor='w', top=False, bottom=False, left=False, right=False)
ax.set_ylabel(r'$\langle\sigma_x\rangle$', fontsize=18, labelpad=30)
ax.set_xlabel(r'${\rm Time}$', fontsize=18, labelpad=10)

rank1, X_test, X_predicted, X_test_noisy = [data[size_of_mem][list_eps[0]][K][key] for key in ['rank', 'X_test', 'X_predicted', 'X_test_noisy']]
time_pred = np.arange(0, (X_predicted.shape[0]) * time_step, time_step) + (K-1) * time_step
time_test = np.arange(0, (X_test.shape[1]) * time_step, time_step)

upper_bound = X_test_noisy[0, :, 0].numpy().max() + 0.12
lower_bound = X_test_noisy[0, :, 0].numpy().min() - 0.12

ax1 = fig.add_subplot(221)
ax1.fill_between((-100 * time_step, (K-1) * time_step), -1, 1, facecolor='red', alpha=0.3)
ax1.fill_between(((K-1) * time_step, (time_test.shape[0] + 100) * time_step), -1, 1, facecolor='green', alpha=0.3)
ax1.set_ylim(top=upper_bound, bottom=lower_bound)
ax1.set_xlim(left=0, right=time_test.shape[0]*time_step)
plt.tick_params(bottom=False, labelbottom=False)
plt.text(22, -0.37, r'$\sigma=$ '+str(list_eps[0]) + ', ' + r'$r=$ '+str(rank1), fontsize=15, bbox=dict(facecolor='white'))
plt.plot(time_test, X_test_noisy[0, :, 0], 'k', alpha=0.2, linewidth=3)
plt.plot(time_test, X_test[0, :, 0], 'b')
plt.plot(time_pred, X_predicted[:, 0], 'r--')
plt.tick_params(axis='both', which='major', labelsize=15)


rank2, X_test, X_predicted, X_test_noisy = [data[size_of_mem][list_eps[1]][K][key] for key in ['rank', 'X_test', 'X_predicted', 'X_test_noisy']]

ax2 = fig.add_subplot(222)
ax2.fill_between((-100 * time_step, (K-1) * time_step), -1, 1, facecolor='red', alpha=0.3)
ax2.fill_between(((K-1) * time_step, (time_test.shape[0] + 100) * time_step), -1, 1, facecolor='green', alpha=0.3)
ax2.set_ylim(top=upper_bound, bottom=lower_bound)
ax2.set_xlim(left=0, right=time_test.shape[0]*time_step)
ax2.set_ylim(top=upper_bound, bottom=lower_bound)
plt.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.text(22, -0.37, r'$\sigma=$ '+str(list_eps[1]) + ', ' + r'$r=$ '+str(rank2), fontsize=15, bbox=dict(facecolor='white'))
plt.plot(time_test, X_test_noisy[0, :, 0], 'k', alpha=0.2, linewidth=3)
plt.plot(time_test, X_test[0, :, 0], 'b')
plt.plot(time_pred, X_predicted[:, 0], 'r--')
plt.tick_params(axis='both', which='major', labelsize=15)

rank3, X_test, X_predicted, X_test_noisy = [data[size_of_mem][list_eps[2]][K][key] for key in ['rank', 'X_test', 'X_predicted', 'X_test_noisy']]

ax3 = fig.add_subplot(223)
ax3.set_ylim(top=upper_bound, bottom=lower_bound)
ax3.fill_between((-100 * time_step, (K-1) * time_step), -1, 1, facecolor='red', alpha=0.3)
ax3.fill_between(((K-1) * time_step, (time_test.shape[0] + 100) * time_step), -1, 1, facecolor='green', alpha=0.3)
ax3.set_ylim(top=upper_bound, bottom=lower_bound)
ax3.set_xlim(left=0, right=time_test.shape[0]*time_step)
plt.text(22, -0.37, r'$\sigma=$ '+str(list_eps[2]) + ', ' + r'$r=$ '+str(rank3), fontsize=15, bbox=dict(facecolor='white'))
plt.plot(time_test, X_test_noisy[0, :, 0], 'k', alpha=0.2, linewidth=3)
plt.plot(time_test, X_test[0, :, 0], 'b')
plt.plot(time_pred, X_predicted[:, 0], 'r--')
plt.tick_params(axis='both', which='major', labelsize=15)

rank4, X_test, X_predicted, X_test_noisy = [data[size_of_mem][list_eps[3]][K][key] for key in ['rank', 'X_test', 'X_predicted', 'X_test_noisy']]

ax4 = fig.add_subplot(224)
ax4.set_ylim(top=upper_bound, bottom=lower_bound)
ax4.fill_between((-100 * time_step, (K-1) * time_step), -1, 1, facecolor='red', alpha=0.3)
ax4.fill_between(((K-1) * time_step, (time_test.shape[0] + 100) * time_step), -1, 1, facecolor='green', alpha=0.3)
ax4.set_ylim(top=upper_bound, bottom=lower_bound)
ax4.set_xlim(left=0, right=time_test.shape[0]*time_step)
plt.tick_params(left=False, labelleft=False)
plt.text(22, -0.37, r'$\sigma=$ '+str(list_eps[3]) + ', ' + r'$r=$ '+str(rank4), fontsize=15, bbox=dict(facecolor='white'))
plt.plot(time_test, X_test_noisy[0, :, 0], 'k', alpha=0.2, linewidth=3)
plt.plot(time_test, X_test[0, :, 0], 'b')
plt.plot(time_pred, X_predicted[:, 0], 'r--')
plt.tick_params(axis='both', which='major', labelsize=15)
plt.subplots_adjust(hspace=0.12, wspace=0.035)

plt.savefig('random_lindbladians_dynamics_prediction.pdf')
```

## Plotting $r$ and $(d_sd_e)^2$ vs level of noise $\sigma$.

```python
# ------------------------Params----------------------- #
list_of_mem_dims = [2, 3, 4, 5, 6]
K = 75
list_of_eps = [0.001, 0.003, 0.01, 0.03, 0.1]
time_step = 0.2
# ----------------------------------------------------- #

# plotting
plt.plot(list_of_mem_dims, [(2 * mem_dim) ** 2 for mem_dim in list_of_mem_dims], '-ob')
styles = ['-or', '-^r', '-Dr', '-sr', '-<r', '-*r']
for eps, style in zip(list_of_eps, styles):
    plt.plot(list_of_mem_dims,
             [data[mem_dim][eps][K]['rank'] for mem_dim in list_of_mem_dims],
             style)
plt.legend([r'$d^2d_{\rm E}^2$'] + list(map(lambda x: r'$r$  for $\sigma=$ ' + str(x), list_of_eps)), fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=15)
plt.xlabel(r'$d_{\rm E}$', fontsize=15)
plt.savefig('rank_vs_sigma.pdf')
```

## Ploting spectrum for different level of noise $\sigma$

```python
# ------------------------Params----------------------- #
mem_dim = 4
K = 75
list_of_eps = [1e-10, 0.01]
time_step = 0.2
# ----------------------------------------------------- #

true_phi_lmbd = data[mem_dim][list_eps[0]][K]['true_phi_lmbd']

# plotting of the spectrum
fig, ax1 = plt.subplots()

# These are in unitless percentages of the figure size. (0,0 is bottom left)
plt.scatter(np.real(true_phi_lmbd),
            np.imag(true_phi_lmbd), color='b', marker='+', s=200)

styles = ['x', '>']
colors = ['r', 'g']

list_of_eps_for_legend = list(map(lambda x: 0 if x < 1e-5 else x, list_of_eps))
for eps, style, color in zip(list_of_eps, styles, colors):
    plt.scatter(np.real(data[mem_dim][eps][K]['learned_phi_lmbd']),
                np.imag(data[mem_dim][eps][K]['learned_phi_lmbd']), marker=style, color=color, s=50)
plt.xlabel(r'${\rm Re}\lambda$', fontsize=15)
plt.ylabel(r'${\rm Im}\lambda$', fontsize=15)
plt.legend([r'Spectrum of $\exp(\tau {\cal L})$'] + list(map(lambda x: r'$\Lambda_r$  for $\sigma=$ ' + str(x), list_of_eps_for_legend)), fontsize=11, loc=[0.485, 0.387])
plt.tick_params(axis='both', which='major', labelsize=15)
left, bottom, width, height = [0.2, 0.3, 0.29, 0.45]
ax2 = fig.add_axes([left, bottom, width, height])
ax2.set_xlim(left=0.98, right=1.002)
ax2.set_ylim(bottom=-0.3, top=0.3)
ax2.scatter(np.real(true_phi_lmbd),
            np.imag(true_phi_lmbd), color='b', marker='+', s=200)
for eps, style, color in zip(list_of_eps, styles, colors):
    ax2.scatter(np.real(data[mem_dim][eps][K]['learned_phi_lmbd']),
                np.imag(data[mem_dim][eps][K]['learned_phi_lmbd']), marker=style, color=color, s=50)
ax2.tick_params(axis='both', which='major', labelsize=12)
plt.savefig('spectrum.pdf')
```

# Plotting prediction error for different level of noise

```python
# ------------------------Params----------------------- #
mem_dim = 4
list_of_K = [5, 15, 30, 45, 60, 75]
list_of_eps = [0.001, 0.003, 0.01, 0.03, 0.1]
time_step = 0.2
# ----------------------------------------------------- #

styles = ['-ob', '-^b', '-Db', '-sb', '-<b', '-*b']
sizes = [6, 6, 6, 6, 8, 10]
for K, style, size in zip(list_of_K, styles, sizes):
    plt.plot(list_of_eps,
             [np.mean(np.linalg.norm(data[mem_dim][eps][K]['X_predicted'] - data[mem_dim][eps][K]['X_test'][:, K-1:-1], axis=-1, ord=2)) for eps in list_of_eps],
             style,
             markersize=size)
plt.xscale('log')
plt.yscale('log')
plt.legend(list(map(lambda x: r'$K=$ ' + str(x), list_of_K)), fontsize=15)
plt.xlabel(r'$\sigma$', fontsize=15)
plt.ylabel(r'${\rm dist}({\cal T}_{\rm predicted}, {\cal T}_{\rm wo})$', fontsize=15)
plt.tick_params(axis='both', which='major', labelsize=15)
plt.gcf().subplots_adjust(left=0.15)
plt.savefig('err_vs_sigma.pdf')
```

```python
# ------------------------Params----------------------- #
list_of_mem_dims = [3, 4, 5, 6]
list_of_K = 60
list_of_eps = [0.001, 0.003, 0.01, 0.03]
time_step = 0.2
# ----------------------------------------------------- #

styles_denoised = ['o-b', 'o--b', 'o:b', 'o-.b', 'o-.b']
styles_noisy = ['o-r', 'o--r', 'o:r', 'o-.r', 'o-.r']
lines1 = []
for i, mem_dim in enumerate(list_of_mem_dims):
    line, = plt.plot(list_of_eps, [np.mean(np.linalg.norm(data[mem_dim][eps][K]['denoised_X'] - data[mem_dim][eps][K]['noiseless_X'], axis=-1)) for eps in list_of_eps], styles_denoised[i])
    lines1.append(line)
legend1 = plt.legend(lines1, [r'${\rm dist}({\cal S}_{\rm wo}, {\cal S}_{\rm denoised})$, $d_{\rm E}=$ ' + str(mem_dim) for mem_dim in list_of_mem_dims], fontsize=12, loc=4)
lines2 = []
for i, mem_dim in enumerate(list_of_mem_dims):
    line, = plt.plot(list_of_eps, [np.mean(np.linalg.norm(data[mem_dim][eps][K]['noisy_X'] - data[mem_dim][eps][K]['noiseless_X'], axis=-1)) for eps in list_of_eps], styles_noisy[i])   
    lines2.append(line)
plt.yscale('log')
plt.xscale('log')
plt.xlabel(r'$\sigma$', fontsize=15)
legend2 = plt.legend(lines2, [r'${\rm dist}({\cal S}_{\rm wo}, {\cal S})$, $d_{\rm E}=$ ' + str(mem_dim) for mem_dim in list_of_mem_dims], fontsize=12, loc=0)
plt.gca().add_artist(legend1)
plt.gca().add_artist(legend2)
plt.tick_params(axis='both', which='major', labelsize=15)
plt.gcf().subplots_adjust(left=0.15)
plt.savefig('denoising.pdf')
```
